#Treating missing values categorical first


```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
```

```python
df['Dependents'] = df['Dependents'].str.replace('+','')     #REPLACING THE '+' SIGN
```

```python
df['Dependents']=df['Dependents'].astype('int')      #CONVERTING DEPENDENTS TO INTEGER DATATYPE
```

```python
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

#Numerical values

```python
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
```

```python
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
```

```python
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].median())
```

```python
df.isna().sum() #missing values treated
```

```python
df.dtypes
```

#Encoding

```python
x = df.drop('Loan_Amount_Term',axis=1)
y = df['Loan_Amount_Term']
```

```python
x.tail()
```

```python
y.head()
```

```python
x = pd.get_dummies(x,dtype =int)             #ONE HOT ENCODING
```

```python
x.head()
```

```python
#LABEL ENCODING
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

```python
y
```

#Scaling

```python
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
```

```python
x.loc[:,['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']] = min_max.fit_transform(x[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']])
```

```python
x.head()
```

#TrainTestSplit


```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)
```

#Importing Regressors RandomForest,XGB,CatBoost

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
!pip install catboost #MANUALLY INSTALLING CATBOOST
```

```python
from catboost import CatBoostRegressor
```

```python
models ={'Random Forest': RandomForestRegressor(),   #IMPORTING THE MODELS TO A VARIABLE
         'XGB Regressor': XGBRegressor(),
         'Catboost': CatBoostRegressor(verbose=0)}
```

```python
for name,model in models.items():
  model.fit(x_train,y_train)
  print(name+'','trained')
```

```python
from sklearn.metrics import mean_squared_error    #CHECKING ACCURACY OF THE MODELS

for name,model in models.items():
  y_pred = model.predict(x_test)
  mse = mean_squared_error(y_test,y_pred)
  print(name +' MSE:{:.4f} '.format(mse))
```

Conclusion is that Random Forest provides the most accurate model for the above dataset among the 3 Regression models used.
